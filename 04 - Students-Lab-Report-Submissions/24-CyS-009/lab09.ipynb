```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("heart_2020_cleaned.csv")
```

```python
age_map = {
    '18-24':21, '25-29':27, '30-34':32, '35-39':37,
    '40-44':42, '45-49':47, '50-54':52, '55-59':57,
    '60-64':62, '65-69':67, '70-74':72, '75-79':77, '80 or older':85
}

df['AgeNum'] = df['AgeCategory'].map(age_map)
print("Mean Age:", df['AgeNum'].mean())
```

```python
print(df['HeartDisease'].value_counts())
```

```python
print("Individuals with BMI > 30:", (df['BMI'] > 30).sum())
```

```python
plt.hist(df['BMI'], bins=20)
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.title("BMI Distribution")
plt.show()
```

```python
df[df['HeartDisease']=='Yes']['PhysicalActivity'].value_counts().plot(kind='bar')
plt.xlabel("Exercise Habit")
plt.ylabel("Heart Disease Cases")
plt.title("Heart Disease by Exercise Habit")
plt.show()
```

```python
risk = df.groupby('PhysicalActivity')['HeartDisease'].value_counts(normalize=True)
print(risk)
```

```python
df['Smoking'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Smokers vs Non-Smokers")
plt.ylabel("")
plt.show()
```

```python
plt.boxplot(df['SleepTime'])
plt.ylabel("Sleep Hours")
plt.title("Sleep Hours Distribution")
plt.show()
```

```python
print("Median Sleep Hours:", df['SleepTime'].median())
```
